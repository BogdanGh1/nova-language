```python
import sqlite3

# Connect to the database (if it doesn't exist, it will be created)
conn = sqlite3.connect('database.db')

# Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Create a table
cursor.execute('''CREATE TABLE IF NOT EXISTS users (
                    id INTEGER PRIMARY KEY,
                    username TEXT,
                    password TEXT
                )''')
```

```python
cursor = conn.cursor()
```

```python
data_to_insert = [
    ('Alice', 30),
    ('Bob', 25),
    ('Charlie', 35)
]
```

```python
cursor.executemany("INSERT INTO users (username, password) VALUES (?, ?)", data_to_insert)
```

```python
cursor.execute("SELECT * FROM users")
```

```python
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)
```

```python
cursor.execute("SELECT * FROM users WHERE username=? and password=?", ("Alice","30"))
results = cursor.fetchall()
```
